```python
# first we will make a fuanction to calculate the mean of values
def Calculate_mean  (data):
  mean = sum (data)/len(data)
  return mean
  # end of function

# our data
data_set = [10, 15, 20, 25, 30]

mean_result = Calculate_mean(data_set)

print (f"Mean resulte = {mean_result}")
```

## 2) median

image.png

```python
def calculate_median (data):
  data.sort() # to sort data
  lengthOfData = len(data)
  mid = lengthOfData // 2 # to get a floor division

  if lengthOfData % 2 == 0 :
    median =(data[mid - 1] + data[mid]) / 2

  else :
    median = data[mid]


  return median
  # end of function


# our data
data_set = [10, 20, 30, 25, 50]

median_result = calculate_median(data_set)

print (f"Median resulte = {median_result}") # medain = 25

# our data
data_set = [10, 15, 20, 25, 30, 40]

median_result = calculate_median(data_set)

print (f"Median resulte = {median_result}") # median = (20+25)/2 = 22.5
```

## 3) mode

image.png

```python
from collections import Counter

def calculate_mode(data):
    data_count = Counter(data)
    max_count = max(data_count.values())
    mode = [k for k, v in data_count.items() if v == max_count]

    if len(mode) == 1:
        return mode[0]
    else:
        return mode  # in case of multiple modes, return all
    # end of function

# Example usage
data_set = [10, 15, 20, 25, 30, 40, 20]

mode_result = calculate_mode(data_set)

print(f"Mode result = {mode_result}")  # mode = 20 (appears twice)
```

# 8) Confusion Matrix

image.png

in 2D

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LogisticRegression

# Load the digits dataset
digits = load_digits()
X, y = digits.data, digits.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train a logistic regression model
model = LogisticRegression(max_iter=10000)
model.fit(X_train, y_train)

# Predict the labels for the test set
y_pred = model.predict(X_test)

# Calculate the confusion matrix
ConfusionMatrix = confusion_matrix(y_test, y_pred)

# Plot the confusion matrix using seaborn's heatmap
plt.figure(figsize=(10, 7))
sns.heatmap(ConfusionMatrix, annot=True, fmt='d', cmap='Blues', xticklabels=digits.target_names, yticklabels=digits.target_names)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

# 10) Hierarchical Clustering

image.png

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_blobs
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage

# Generate sample data
X, y = make_blobs(n_samples=300, centers=4, random_state=42)

# Compute the linkage matrix
linkage_matrix = linkage(X, method='ward')

# Plot the dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linkage_matrix)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample Index')
plt.ylabel('Distance')
plt.show()

# Perform Agglomerative Clustering
clusters = AgglomerativeClustering(n_clusters=4, metric='euclidean', linkage='ward')
y_pred = clusters.fit_predict(X)

# Plot the clustered data
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
plt.title('Agglomerative Hierarchical Clustering')
plt.show()
```

# 12) Cross Validation

image.png

```python
from sklearn.datasets import load_iris # dataset
from sklearn.model_selection import KFold, cross_val_score
from sklearn.svm import SVC

iris = load_iris()
X, y = iris.data, iris. target

svm_classifier = SVC(kernel='linear', C=1)
kfold = KFold(n_splits=5, shuffle=True, random_state=42)
accuracy_scores = cross_val_score(svm_classifier, X, y, cv=kfold)

print("Accuracy Scores for each fold: ", accuracy_scores)
print("Mean Accuracy: ", accuracy_scores.mean())
```
